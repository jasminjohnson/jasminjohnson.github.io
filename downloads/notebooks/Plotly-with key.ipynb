```python
# ref: https://plot.ly/python/ipython-notebook-tutorial/#import-data


# Import the file
df = pd.read_csv('data/LifePlotly.csv')

# You can import the file from sepecific website, for example if store the file in github
# df = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/LifePlotly.csv")

df.head()

# we can see the country name for 200th
country = df.CountryName
country[200]

# Calling std() calculates the standard deviation for each column.
df.std()

# Create a table
table = ff.create_table(df)


# View table online
# py.iplot(table, filename='jupyter-table1')  # create table online

# View table 
# plotly.offline.iplot(table, filename='jupyter-table1')   
py.iplot(table, filename='jupyter-table1')
```

#### *Plot Dashboard online or offline
- We can view the dashboard online or offline
- Dashboard in Bar graph
- Dashboard in Scatter graph with title

```python
import plotly.plotly as py
import plotly.graph_objs as go

data = [go.Bar(x=df.CountryName,
            y=df.Birthrate)]

# View dashboard online
py.iplot(data, filename='jupyter-basic_bar')

# View dashboard offline
plotly.offline.iplot(data, filename='jupyter-basic_bar') 
```

```python
import plotly
import plotly.graph_objs as go

# plotly.offline.init_notebook_mode(connected=True)

# View online
py.iplot({
    "data": [go.Scatter(x= df.CountryName, y= df.Birthrate)],
    "layout": go.Layout(title="World Birthrate")
})


# View offline
plotly.offline.iplot({
    "data": [go.Scatter(x= df.CountryName, y= df.Birthrate)],
    "layout": go.Layout(title="World Birthrate")
})
```

#### *View dashboard in Plotly *

```python
# Birthrate = df.Birthrate
# Deathrate = df.Deathrate


trace0 = go.Scatter(
    x= df.CountryName,
    y= df.Birthrate
)
trace1 = go.Scatter(
    x= df.CountryName,
    y= df.Population
)
data = [trace0, trace1]

# View online
py.iplot(data, filename = 'basic-line', auto_open=True)

# View offline
plotly.offline.iplot(data, filename = 'basic-line')

```
